```python
import pandas as pd
df = pd.read_csv('mock_kaggle.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['data'] = pd.to_datetime(df['data'])
df.info()
```

```python
import matplotlib.pyplot as plt

# Daily sales trend
plt.figure(figsize=(10,5))
plt.plot(df['data'], df['venda'])
plt.title("Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales (venda)")
plt.show()
```

```python
# Month column create
df['month'] = df['data'].dt.month

# Monthly total sales
monthly_sales = df.groupby('month')['venda'].sum()
monthly_sales
```

```python
# Monthly Sales Bar Chart
plt.figure(figsize=(10,5))
plt.bar(monthly_sales.index, monthly_sales.values)
plt.title("Monthly Sales Analysis")
plt.xlabel("Month")
plt.ylabel("Total Sales (venda)")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df['estoque'], df['venda'])
plt.title("Stock vs Sales")
plt.xlabel("Stock (estoque)")
plt.ylabel("Sales (venda)")
plt.show()
```
